```python
from sklearn.cluster import KMeans
```

```python
df.drop([
'sepal length (cm)','sepal width (cm)'],axis=1,inplace=True)
```

```python
df
```

```python
model=KMeans(n_clusters=3)
```

```python
y_pred=model.fit_predict(df)
```

```python
y_pred
```

```python
df['cluster']=y_pred
```

```python
model.cluster_centers_
```

```python
r1=df[df.cluster==0]
r2=df[df.cluster==1]
r3=df[df.cluster==2]
```
